**1. Level of measurement(scale) of energy feature is ratio scale**

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io.wavfile
import scipy.stats as pdf

# 3.a) load files

train_sample_rate, train_samples = scipy.io.wavfile.read('TrainingFile.wav')
test_sample_rate, test_samples = scipy.io.wavfile.read('TestingFile.wav')

# 3.b) Normalize
# Divide positive values with max limit of int16 - 32767
# Divide negative values with min limit of int16 - 32768

#For train_samples
max_number = np.amax(train_samples)
min_number = np.amin(train_samples)

train_samples = train_samples.astype('float64')

for i in range(len(train_samples)):
    if np.any(train_samples[i] > 0):
        train_samples[i] = train_samples[i] / 32767 #max_number
    else:
        train_samples[i] = train_samples[i] / 32768 #min_number

#For test_samples
max_number = np.amax(test_samples)
min_number = np.amin(test_samples)

test_samples = test_samples.astype('float64')

for i in range(len(test_samples)):
    if np.any(test_samples[i] > 0):
        test_samples[i] = test_samples[i] / 32767 #max_number
    else:
        test_samples[i] = test_samples[i] / 32768 #min_number
```

```python
# 4. Calculating energy between interval 0..len(train_samples) (in milliseconds unit) with window = 0.01 * rate

train_sample_duration = len(train_samples) / train_sample_rate
test_sample_duration = len(test_samples) / test_sample_rate

def calculate_energy(duration, rate, samples):
    #Converting window value(10 milliseconds) to its equivalent number of samples
    energy = np.zeros(int(duration * 1000))
    for i in range(0, len(energy)):
        starting_sample_index = int(rate * i / 1000)
        window_samples = int(0.01 * rate)
        for j in range(starting_sample_index, min(window_samples + starting_sample_index, len(samples))):
            energy[i] += samples[j]**2
    return energy

train_energy = calculate_energy(train_sample_duration, train_sample_rate, train_samples)
test_energy = calculate_energy(test_sample_duration, test_sample_rate, test_samples)
```

```python
# 5. Silence region

# 5.a) Labeling train_energy with silence or voice
threshold = 0.15
train_samples_x  = np.arange(0, len(train_energy)) / 1000
train_energy_masked_voice = np.ma.masked_greater(train_energy, threshold)
train_energy_masked_silence = np.ma.masked_less(train_energy, threshold)

# 5.b) Boolean array for silence and voice regions
train_attribute = np.zeros(len(train_energy), bool)
train_attribute[train_energy <= threshold] = 1

# 5.c) Prior probabilities - voice and silence
train_count_silence = 0
train_count_voice = 0
train_voice_energy = np.array([])
train_silence_energy = np.array([])
for i in range(0, len(train_attribute)):
    if train_attribute[i] == 1:
        train_count_silence += 1
        train_silence_energy = np.append(train_silence_energy, train_energy[i])
    else:
        train_count_voice += 1
        train_voice_energy = np.append(train_voice_energy, train_energy[i])

prior_silence = train_count_silence / len(train_energy)
prior_voice = train_count_voice / len(train_energy)

# 5.d) Statistical distributions of silence and voice
distributions_silence = pdf.norm(loc = np.mean(train_silence_energy), scale = np.std(train_silence_energy))
distributions_voice = pdf.norm(loc = np.mean(train_voice_energy), scale = np.std(train_voice_energy))
```

```python
# 6. storing the scaled likelihood and classifiying each value as either silence(1) or voice(0)
test_samples_x = np.arange(0, len(test_energy)) / 1000
test_classification = np.zeros(len(test_energy), bool)
scaled_likelihood_silence = np.zeros(len(test_energy), float)
scaled_likelihood_voice = np.zeros(len(test_energy), float)
for i in range(len(test_classification)):
    scaled_likelihood_silence[i] = distributions_silence.pdf(test_energy[i]) * prior_silence
    scaled_likelihood_voice[i] = distributions_voice.pdf(test_energy[i]) * prior_voice
    if np.any(scaled_likelihood_silence[i] > scaled_likelihood_voice[i]):
        test_classification[i] = 1
    else:
        test_classification[i] = 0

#Masking based on the classification
test_masked_silence = np.ma.masked_where(test_classification == 1, test_energy)
test_masked_voice = np.ma.masked_where(test_classification == 0, test_energy)
```

```python
# 7. Plotting energy values, distributions and scaled likelihoods

plt.plot(train_samples_x, train_energy_masked_silence, train_samples_x, train_energy_masked_voice)
plt.xlabel("Time [s]")
plt.ylabel("Energy")
plt.title("Train")
plt.legend(("Voice", "Silence"))
plt.show()


plt.plot(test_samples_x, test_masked_silence, test_samples_x, test_masked_voice)
plt.xlabel("Time [s]")
plt.ylabel("Energy")
plt.title("Test")
plt.legend(("Voice", "Silence"))
plt.show()

x = np.linspace(-2, 4, 1000)
plt.plot(x, distributions_voice.pdf(x), x, distributions_silence.pdf(x))
plt.xlabel("x")
plt.ylabel("$P(x|\omega)$")
plt.title("Gamma/Normal distributions")
plt.legend(("$N(Voice)$", "$N(Silence)$"))
plt.show()

plt.plot(test_samples_x, scaled_likelihood_voice, test_samples_x, scaled_likelihood_silence)
plt.xlabel("Time [s]")
plt.ylabel("$p(x|\omega) \cdot P(\omega)$")
plt.title("Test")
plt.legend(("$p(x|Voice) \cdot P(Voice)$", "$p(x|Silence) \cdot P(Silence)$"), loc = "center right")
plt.show()
```
